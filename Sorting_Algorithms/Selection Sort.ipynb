```python
def selection_sort(arr,n):
    cmp=0
    swaps=0
    for i in range(n):
        min_idx=i
        for j in range(i+1,n):
            cmp+=1
            if(arr[min_idx]>arr[j]):
                min_idx=j
        swaps+=1        
        temp=arr[min_idx]
        arr[min_idx]=arr[i]
        arr[i]=temp
    return cmp,swaps
```

```python
import time
def selection_sort_timer(arr,n):
    start_time=time.time()
    for i in range(n):
        min_idx=i
        for j in range(i+1,n):
            
            if(arr[min_idx]>arr[j]):
                min_idx=j
        temp=arr[min_idx]
        arr[min_idx]=arr[i]
        arr[i]=temp
         
    return (time.time()-start_time)
```

```python
inp_sizes = [10,50,100,500,1000,5000,10000]
import random as rnd
import matplotlib.pyplot as plt
```

```python
ss_s=[]
ss_c=[]
ss_t=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    cmps,swps=selection_sort(arr,n)
    ts=selection_sort_timer(arr,n)
    ss_s.append(swps)
    ss_c.append(cmps)
    ss_t.append(ts)
print("Comparisons:",ss_c)
print("Swaps:",ss_s)
print("Execution time:",ss_t)
```

```python
p1=plt.plot(inp_sizes,ss_c)
plt.xlabel("Input sizes")
plt.ylabel("No. of comparisons")
plt.show()
```

```python
#Testing the algorithm on a "nearly sorted array"
ss_s1=[]
ss_c1=[]
ss_t1=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    arr.sort()
    cmps,swps=selection_sort(arr,n)
    ts=selection_sort_timer(arr,n)
    ss_s1.append(swps)
    ss_c1.append(cmps)
    ss_t1.append(ts)
print("Comparisons:",ss_c1)
print("Swaps:",ss_s1)
print("Execution time:",ss_t1)
```

```python
#Testing the algorithm on a "reverse sorted" array
ss_s2=[]
ss_c2=[]
ss_t2=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    arr.sort(reverse=True)
    cmps,swps=selection_sort(arr,n)
    ts=selection_sort_timer(arr,n)
    ss_s2.append(swps)
    ss_c2.append(cmps)
    ss_t2.append(ts)
print("Comparisons:",ss_c2)
print("Swaps:",ss_s2)
print("Execution time:",ss_t2)
```

```python
#Graph to analyse number of swaps made by algorithm for different nature of array
plt.plot(inp_sizes,ss_s)
plt.plot(inp_sizes,ss_s1)
plt.plot(inp_sizes,ss_s2)
plt.xlabel("Input Sizes")
plt.ylabel("No. of swaps")
plt.legend(["Random array","Nearly sorted array","Reverse sorted array"])
```

```python
#Graph to analyse execution time made by algorithm for different nature of array
plt.plot(inp_sizes,ss_t)
plt.plot(inp_sizes,ss_t1)
plt.plot(inp_sizes,ss_t2)
plt.xlabel("Input Sizes")
plt.ylabel("Execution Time")
plt.legend(["Random array","Nearly sorted array","Reverse sorted array"])
```
