```python
import random as rnd
import numpy
import matplotlib.pyplot as plt
import time
import math
```

```python
#Preparing list of array sizes
arr_sizes = []
for i in range(1,11):
    arr_sizes.append(100*i)    
```

```python
#Bubble sort
def bubble_sort(arr,n):
    i=0
    while(i<n):
        j=0
        while(j<n-1-i):
            if(arr[j]>arr[j+1]):
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
            j+=1
        i+=1

#Arrays to store runtimes for different array sizes and different types of sorted arrays
random_times=[]
sorted_times=[]
part_sorted_times=[]
rev_sorted_times=[]
for n in arr_sizes:
    random_array=[]
    for i in range(n):
        random_array.append(rnd.randint(-1000,1000))
    
    #Random array
    start=time.time()
    bubble_sort(random_array,n)
    end=time.time()
    random_times.append(end-start) 
    
    #Reverse Sorted array (descending)
    random_array.sort(reverse=True)
    start=time.time()
    bubble_sort(random_array,n)
    end=time.time()
    rev_sorted_times.append(end-start)
    
    #Sorted array
    random_array.sort()
    start=time.time()
    bubble_sort(random_array,n)
    end=time.time()
    sorted_times.append(end-start)
    
    #Partially sorted array
    d=5
    l=0
    r=d
    part_sorted_array=[]
    while(r<=n):
        temp=random_array[l:r:1]
        rnd.shuffle(temp)
        for i in range(len(temp)):
            part_sorted_array.append(temp[i])
        l=r
        r+=d
    start=time.time()
    bubble_sort(random_array,n)
    end=time.time()
    part_sorted_times.append(end-start)
    
#Plotting
plt.plot(arr_sizes,random_times)
plt.plot(arr_sizes,sorted_times)
plt.plot(arr_sizes,rev_sorted_times)
plt.plot(arr_sizes,part_sorted_times)
plt.xlabel("Array sizes")
plt.ylabel("Runtimes")
plt.legend(["Random","Sorted","Reverse sorted","Partially sorted"])
plt.title("Bubble Sort")


    
    
    
```

```python
#Insertion sort

def insertion_sort(arr,n):
    i=0
    while(i<n):
        j=i
        while(j>0):
            if(arr[j]<arr[j-1]):
                temp=arr[j]
                arr[j]=arr[j-1]
                arr[j-1]=temp
            j-=1
        i+=1

#Arrays to store runtimes for different array sizes and different types of sorted arrays
random_times=[]
sorted_times=[]
part_sorted_times=[]
rev_sorted_times=[]
for n in arr_sizes:
    random_array=[]
    for i in range(n):
        random_array.append(rnd.randint(-1000,1000))
    
    #Random array
    start=time.time()
    insertion_sort(random_array,n)
    end=time.time()
    random_times.append(end-start) 
    
    #Reverse Sorted array (descending)
    random_array.sort(reverse=True)
    start=time.time()
    insertion_sort(random_array,n)
    end=time.time()
    rev_sorted_times.append(end-start)
    
    #Sorted array
    random_array.sort()
    start=time.time()
    insertion_sort(random_array,n)
    end=time.time()
    sorted_times.append(end-start)
    
    #Partially sorted array
    d=5
    l=0
    r=d
    part_sorted_array=[]
    while(r<=n):
        temp=random_array[l:r:1]
        rnd.shuffle(temp)
        for i in range(len(temp)):
            part_sorted_array.append(temp[i])
        l=r
        r+=d
    start=time.time()
    insertion_sort(random_array,n)
    end=time.time()
    part_sorted_times.append(end-start)
    
#Plotting
plt.plot(arr_sizes,random_times) 
plt.plot(arr_sizes,sorted_times)
plt.plot(arr_sizes,rev_sorted_times)
plt.plot(arr_sizes,part_sorted_times)
plt.xlabel("Array sizes")
plt.ylabel("Runtimes")
plt.legend(["Random","Sorted","Reverse sorted","Partially sorted"])
plt.title("Insertion Sort")


    
    
    
```

```python
#Selection Sort
def selection_sort(arr,n):

    for i in range(n):
        min_idx=i
        for j in range(i+1,n):

            if(arr[min_idx]>arr[j]):
                min_idx=j
      
        temp=arr[min_idx]
        arr[min_idx]=arr[i]
        arr[i]=temp

#Arrays to store runtimes for different array sizes and different types of sorted arrays
random_times=[]
sorted_times=[]
part_sorted_times=[]
rev_sorted_times=[]
for n in arr_sizes:
    random_array=[]
    for i in range(n):
        random_array.append(rnd.randint(-1000,1000))
    
    #Random array
    start=time.time()
    selection_sort(random_array,n)
    end=time.time()
    random_times.append(end-start) 
    
    #Reverse Sorted array (descending)
    random_array.sort(reverse=True)
    start=time.time()
    selection_sort(random_array,n)
    end=time.time()
    rev_sorted_times.append(end-start)
    
    #Sorted array
    random_array.sort()
    start=time.time()
    selection_sort(random_array,n)
    end=time.time()
    sorted_times.append(end-start)
    
    #Partially sorted array
    d=5
    l=0
    r=d
    part_sorted_array=[]
    while(r<=n):
        temp=random_array[l:r:1]
        rnd.shuffle(temp)
        for i in range(len(temp)):
            part_sorted_array.append(temp[i])
        l=r
        r+=d
    start=time.time()
    selection_sort(random_array,n)
    end=time.time()
    part_sorted_times.append(end-start)
    
#Plotting
plt.plot(arr_sizes,random_times)
plt.plot(arr_sizes,sorted_times)
plt.plot(arr_sizes,rev_sorted_times)
plt.plot(arr_sizes,part_sorted_times)
plt.xlabel("Array sizes")
plt.ylabel("Runtimes")
plt.legend(["Random","Sorted","Reverse sorted","Partially sorted"])
plt.title("Selection Sort")
```

```python
#Merge Sort
arr_sizes = []
for i in range(1,51):
    arr_sizes.append(200*i)  
def cmbn(A,B): #Function to combine two sequences(sorted) into a single sorted sequence

    C=[]
    while(len(A)>0 or len(B)>0):
        if(len(A)==0 and len(B)>0):
            C.append(B[0])
            B.pop(0)
            
        if(len(B)==0 and len(A)>0):
            C.append(A[0])
            A.pop(0)
            

        if(len(A)>0 and len(B)>0 and A[0]<B[0]): #The smallest between the first elements of A and B is appended to C
            C.append(A[0]) #The element appended is deleted from original array passed to the function
            A.pop(0)
        else:
            if(len(B)>0):

                C.append(B[0])
                B.pop(0)
    return C


def merge_sort(A,l,r):
    if(l<=r):

        if(l==r):
            C=[]
            C.append(A[l])
            return C
        
        if((l==r-1) and r<len(A)):
            C=[]
            C.append(min(A[l],A[r]))
            C.append(max(A[l],A[r]))
            return C
        
        m=math.floor(l+(r-l+1)/2)
        #Now the array is divided into two halves, the halves are sorted individually and merged into one single sorted array
        if(m==r):
            B=merge_sort(A,l,m-1) #left half
        else:
            B=merge_sort(A,l,m) 
        C=merge_sort(A,m+1,r)  #right half 
        D=cmbn(B,C) #Calling the combining function
        return D
    
#Arrays to store runtimes for different array sizes and different types of sorted arrays
random_times=[]
sorted_times=[]
part_sorted_times=[]
rev_sorted_times=[]
for n in arr_sizes:
    random_array=[]
    for i in range(n):
        random_array.append(rnd.randint(-1000,1000))
    
    #Random array
    start=time.time()
    merge_sort(random_array,0,n-1)
    end=time.time()
    random_times.append(end-start) 
    
    #Reverse Sorted array (descending)
    random_array.sort(reverse=True)
    start=time.time()
    merge_sort(random_array,0,n-1)
    end=time.time()
    rev_sorted_times.append(end-start)
    
    #Sorted array
    random_array.sort()
    start=time.time()
    merge_sort(random_array,0,n-1)
    end=time.time()
    sorted_times.append(end-start)
    
    #Partially sorted array
    d=5
    l=0
    r=d
    part_sorted_array=[]
    while(r<=n):
        temp=random_array[l:r:1]
        rnd.shuffle(temp)
        for i in range(len(temp)):
            part_sorted_array.append(temp[i])
        l=r
        r+=d
    start=time.time()
    merge_sort(random_array,0,n-1)
    end=time.time()
    part_sorted_times.append(end-start)
    
#Plotting
plt.plot(arr_sizes,random_times)
plt.plot(arr_sizes,sorted_times)
plt.plot(arr_sizes,rev_sorted_times)
plt.plot(arr_sizes,part_sorted_times)
plt.xlabel("Array sizes")
plt.ylabel("Runtimes")
plt.legend(["Random","Sorted","Reverse sorted","Partially sorted"])
plt.title("Merge Sort")
```

```python
#Quick Sort
arr_sizes = []
for i in range(1,11):
    arr_sizes.append(100*i)

def partition(arr,l,r):
    pivot = arr[r] #last element is pivot
    i=l-1
    for j in range(l,r):
        if(arr[j]<=pivot): #if element is less than pivot, we increase i and swap arr[i] with arr[j]
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[r]=arr[r],arr[i+1] #finally we place pivot at the right position
    return (i+1)

def quick_sort(arr,l,r):
    if(l==r):
        return arr[l]
    if(l<r):
        ind_part = partition(arr,l,r)
        if(ind_part>0):
            quick_sort(arr,l,ind_part-1)
        if(ind_part<(len(arr)-1)):
            quick_sort(arr,ind_part+1,r)
        

#Arrays to store runtimes for different array sizes and different types of sorted arrays
random_times=[]
sorted_times=[]
part_sorted_times=[]
rev_sorted_times=[]
for n in arr_sizes:
    random_array=[]
    for i in range(n):
        random_array.append(rnd.randint(-1000,1000))
    
    #Random array
    start=time.time()
    quick_sort(random_array,0,n-1)
    end=time.time()
    random_times.append(end-start) 
    
    #Reverse Sorted array (descending)
    random_array.sort(reverse=True)
    start=time.time()
    quick_sort(random_array,0,n-1)
    end=time.time()
    rev_sorted_times.append(end-start)
    
    #Sorted array
    random_array.sort()
    start=time.time()
    quick_sort(random_array,0,n-1)
    end=time.time()
    sorted_times.append(end-start)
    
    #Partially sorted array
    d=5
    l=0
    r=d
    part_sorted_array=[]
    while(r<=n):
        temp=random_array[l:r:1]
        rnd.shuffle(temp)
        for i in range(len(temp)):
            part_sorted_array.append(temp[i])
        l=r
        r+=d
    start=time.time()
    quick_sort(random_array,0,n-1)
    end=time.time()
    part_sorted_times.append(end-start)
    
#Plotting
plt.plot(arr_sizes,random_times)
plt.plot(arr_sizes,sorted_times)
plt.plot(arr_sizes,rev_sorted_times)
plt.plot(arr_sizes,part_sorted_times)
plt.xlabel("Array sizes")
plt.ylabel("Runtimes")
plt.legend(["Random","Sorted","Reverse sorted","Partially sorted"])
plt.title("Quick Sort")
```

CONCLUSIONS:

1. As expected the insertion sort and bubble sort show improvement in runtimes as we approach a sorted array

2. Selection sort does not improve runtimes for any type of array

3. Merge sort improves runtimes as we approach a sorted array

4. Quick sort on the other hand (with the last element as pivot), performs best for a random array. Its runtime worsens in the case of a reverse sorted array and it takes even more time when the array is almost sorted or completely sorted, since it performs many swaps.
