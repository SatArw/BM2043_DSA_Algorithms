Given functions can be categorized into 4 groups based on their behaviour as n->infinity.


```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt
from scipy.special import gamma
```

Group A: 
1. n!

2. exp(n)

3. 2^n

4. (3/2)^n

5. n^n

Properties:

a) All these functions are increasing functions.

b) They increase very rapidly as n->infinity.

Note: Order of magnitude among these functions for large n goes like this

n^n > n! > exp(n) > 2^n > (3/2)^n


```python
#y=exp(n)
xa=np.linspace(0,20,100)
z1=np.exp(xa)
plt.plot(xa,z1)
```

```python
#y=2^n
z2=np.power(2,xa)
plt.plot(xa,z2)
```

```python
#y=1.5^n
z3=np.power(1.5,xa)
plt.plot(xa,z3)
```

```python
#y=factorial(n)
#plotted for small n, notice the rapid increase for large n
xf=np.linspace(0,10,1000)
plt.plot(xf,gamma(xf),label='Factorial')
```

```python
#y=n^n
#This function reaches immeasurable heights as n grows.
xn=np.linspace(0,10,1000)
z4=np.power(xn,xn)
plt.plot(xn,z4)
```

Group B:

1. n
2. n^2
3. n^3
4. n^1/3
5. n^1/2
6. 2^logn
7. nlogn

Properties:
a) These functions are increasing.

b) They increase fast when n grows larger but the magnitude of these functions is not as big as group A functions.

Note: Order of magnitude among these functions for large n goes like this

n^3 > n^2 > nlogn > n > n^1/2 > n^1/3 > 2^logn


```python
#y=x
xb=np.linspace(0,100,100)
y1=xb

plt.plot(xb,y1)
```

```python
# y=x^2
y2=np.power(xb,2)
plt.plot(xb,y2)
```

```python
#y=x^3
y3=np.power(xb,3)
plt.plot(xb,y3)
```

```python
#y=n^1/3
y4=np.power(xb,1/3)
plt.plot(xb,y4)
```

```python
#y=n^1/2
y5=np.power(xb,1/2)
plt.plot(xb,y5)
```

```python
#y=2^logn
x_for_log = np.linspace(0.1,100,100)
y6=np.power(2,np.log10(x_for_log))
plt.plot(x_for_log,y6)
```

```python
#y=nlogn
y7=x_for_log * np.log10(x_for_log)
plt.plot(x_for_log,y7)
```

Group C:

1. logn
2. ln(n)
3. log(log(n))
4. ln(ln(n))

Properties:

a) These are logarithmic functions, they are negative up to a certain "N" and become positive for n>N.

b) They are increasing functions but they increase at a very slow pace as n->infinity.

c) These functions are said to be "stable" since they don't "blow up" as n increases.

Note:
Order of magnitude among these functions for large n goes like this

ln(n) > logn > ln(ln(n)) > log(logn) 


```python
#y=logn
y8 = np.log10(x_for_log)
plt.plot(x_for_log,y8)
```

```python
#y=ln(n)
y9=np.log(x_for_log)
plt.plot(x_for_log,y9)
```

```python
#y=log(log(n))
y10=np.log10(np.log10(x_for_log))
plt.plot(x_for_log,y10)
```

```python
#y=ln(ln(n))
y11=np.log(np.log(x_for_log))
plt.plot(x_for_log,y11)
```

Group D:

1. 1/n
2. logn/n
3. ln(n)/n

Properties:

a) The similarity among these functions is that they are decreasing functions (as n->infinity).

b) logn/n and ln(n)/n increase for small n till they reach 0 at n=1. After a certain point both these functions start decreasing.

c) All the three functions converge to 0.


Note:
Order of magnitude among these functions for large n goes like this

ln(n)/n > logn/n > 1/n

```python
# y= 1/n
xd=np.linspace(0.1,100,100)
y12=1/xd
plt.plot(xd,y12)
```

```python
#y=logn/n
y13=np.log10(xd)/xd
plt.plot(xd,y13)
```

```python
#y=ln(n)/n
y14=np.log(xd)/xd
plt.plot(xd,y14)
```
