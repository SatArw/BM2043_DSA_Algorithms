```python
def bubble_sort(arr):
    cmps = 0
    l = len(arr)
    for i in range(l):
        for j in range(0,l-1):
            cmps+=1
            if(arr[j]>arr[j+1]):
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr,cmps

def rev_sort(arr):
    cmps = 0
    l = len(arr)
    for i in range(l):
        for j in range(0,l-1):
            cmps+=1
            if(arr[j]<arr[j+1]):
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr,cmps
def spec_sort(arr):
    comps=0
    l = len(arr)
    incr_sorted_arr,cmps = bubble_sort(arr)
    comps += cmps
    incr_sorted_arr[int(l/2)], incr_sorted_arr[l-1] = incr_sorted_arr[l-1], incr_sorted_arr[int(l/2)]
    decr_sorted_arr,cmps = rev_sort(arr[int(l/2)+1:l:1])
    comps += cmps
    f_arr = incr_sorted_arr[0:int(l/2)+1:1] + decr_sorted_arr
    return f_arr
```

```python
import random as rnd
size = 1000

arr = []
for i in range(size):
    arr.append(rnd.randint(0,1000))
k = 5
k_matrix = [[] for i in range(k+1)]
for j in range(len(arr)):
    rem = int(arr[j])%k
    if(rem==0):
        k_matrix[k].append(arr[j])
    else:
        k_matrix[rem].append(arr[j])
#print(k_matrix)
for i in range(1,k+1):
    this_list = k_matrix[i]
    #print(k_matrix[i])
    this_list = spec_sort(this_list)
    k_matrix[i] = this_list
#print(k_matrix)
final_plot = []
for i in range(1,k+1):
    this_list = k_matrix[i]
    for j in range(len(this_list)):
        final_plot.append(this_list[j])
#print(final_plot)

x = [i for i in range(1,len(arr)+1)]
plt.plot(x,final_plot)
```

Time Complexity

•	There will be k reminder lists. 

•	For every reminder list, I am performing the wave sort on it, which takes O((n/k)^2)

•	For K lists this comes out to be = O(n^2/k)

- If merge sort is used, the complexity can be improved to O(Nlog(N/k))
