```python
def insertion_sort(arr,n):
    i=0
    cmp=0
    swaps=0
    while(i<n):
        j=i
        while(j>0):
            cmp+=1
            if(arr[j]<arr[j-1]):
                swaps+=1
                temp=arr[j]
                arr[j]=arr[j-1]
                arr[j-1]=temp
            j-=1
        i+=1
    return cmp,swaps
```

```python
import time
def insertion_sort_timer(arr,n):
    start_time=time.time()
    i=0
    while(i<n):
        j=i
        while(j>0):
            if(arr[j]<arr[j-1]):
                temp=arr[j]
                arr[j]=arr[j-1]
                arr[j-1]=temp
            j-=1
        i+=1
    return (time.time()-start_time)
```

```python
inp_sizes = [10,50,100,500,1000,5000,10000]
import random as rnd
import matplotlib.pyplot as plt
```

```python
#insertion sort
is_s=[]
is_c=[]
is_t=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    cmps,swps=insertion_sort(arr,n)
    ts=insertion_sort_timer(arr,n)
    is_s.append(swps)
    is_c.append(cmps)
    is_t.append(ts)
print("Comparisons:",is_c)
print("Swaps:",is_s)
print("Execution time:",is_t)  
        
```

```python
#Testing the algorithm on a "nearly sorted array"
is_s1=[]
is_c1=[]
is_t1=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    arr.sort()
    cmps,swps=insertion_sort(arr,n)
    ts=insertion_sort_timer(arr,n)
    is_s1.append(swps)
    is_c1.append(cmps)
    is_t1.append(ts)
print("Comparisons:",is_c1)
print("Swaps:",is_s1)
print("Execution time:",is_t1)
```

```python
#Testing the algorithm on a "reverse sorted" array
is_s2=[]
is_c2=[]
is_t2=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    arr.sort(reverse=True)
    cmps,swps=insertion_sort(arr,n)
    ts=insertion_sort_timer(arr,n)
    is_s2.append(swps)
    is_c2.append(cmps)
    is_t2.append(ts)
print("Comparisons:",is_c2)
print("Swaps:",is_s2)
print("Execution time:",is_t2)
```

```python
#Graph to analyse number of swaps made by algorithm for different nature of array
plt.plot(inp_sizes,is_s)
plt.plot(inp_sizes,is_s1)
plt.plot(inp_sizes,is_s2)
plt.xlabel("Input Sizes")
plt.ylabel("No. of swaps")
plt.legend(["Random array","Nearly sorted array","Reverse sorted array"])
```

```python
#Graph to analyse execution time made by algorithm for different nature of array
plt.plot(inp_sizes,is_t)
plt.plot(inp_sizes,is_t1)
plt.plot(inp_sizes,is_t2)
plt.xlabel("Input Sizes")
plt.ylabel("Execution Time")
plt.legend(["Random array","Nearly sorted array","Reverse sorted array"])
```
