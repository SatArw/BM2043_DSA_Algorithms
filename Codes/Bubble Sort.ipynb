```python
def bubble_sort(arr,n):
    start_time = time.time()
    i=0
    cmp=0
    swaps=0
    while(i<n):
        j=0
        while(j<n-1-i):
            cmp+=1
            if(arr[j]>arr[j+1]):
                swaps+=1
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
            j+=1
        i+=1
    return cmp,swaps
```

```python
import time
def bubble_sort_timer(arr,n):
    start_time = time.time()
    i=0
    while(i<n):
        j=0
        while(j<n-1-i):
            if(arr[j]>arr[j+1]):
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
            j+=1
        i+=1
    return (time.time()-start_time)
```

```python
inp_sizes = [10,50,100,500,1000,5000,10000]
import random as rnd
import matplotlib.pyplot as plt
```

```python
bs_s=[]
bs_c=[]
bs_t=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    cmps,swps=bubble_sort(arr,n)
    ts=bubble_sort_timer(arr,n)
    bs_s.append(swps)
    bs_c.append(cmps)
    bs_t.append(ts)
print("Comparisons:",bs_c)
print("Swaps:",bs_s)
print("Time Taken for execution:",bs_t)
```

```python
p1=plt.plot(inp_sizes,bs_c)
plt.xlabel("Input sizes")
plt.ylabel("No. of comparisons")
plt.show()
```

```python
#Testing on nearly sorted array
bs_s1=[]
bs_c1=[]
bs_t1=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    arr.sort()
    cmps,swps=bubble_sort(arr,n)
    ts=bubble_sort_timer(arr,n)
    bs_s1.append(swps)
    bs_c1.append(cmps)
    bs_t1.append(ts)
print("Comparisons:",bs_c1)
print("Swaps:",bs_s1)
print("Execution Time:",bs_t1)
```

```python
#Testing on reverse sorted array
bs_s2=[]
bs_c2=[]
bs_t2=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(0,1000000))
    arr.sort(reverse=True)
    cmps,swps=bubble_sort(arr,n)
    ts=bubble_sort_timer(arr,n)
    bs_s2.append(swps)
    bs_c2.append(cmps)
    bs_t2.append(ts)
print("Comparisons:",bs_c2)
print("Swaps:",bs_s2)
print("Execution Time:",bs_t2)
```

```python
#Graph to analyse number of swaps made by algorithm for different nature of array
plt.plot(inp_sizes,bs_s)
plt.plot(inp_sizes,bs_s1)
plt.plot(inp_sizes,bs_s2)
plt.xlabel("Input Sizes")
plt.ylabel("No. of swaps")
plt.legend(["Random array","Nearly sorted array","Reverse sorted array"])

```

```python
#Graph to analyse execution time made by algorithm for different nature of array
plt.plot(inp_sizes,bs_t)
plt.plot(inp_sizes,bs_t1)
plt.plot(inp_sizes,bs_t2)
plt.xlabel("Input Sizes")
plt.ylabel("Execution Time")
plt.legend(["Random array","Nearly sorted array","Reverse sorted array"])
```
