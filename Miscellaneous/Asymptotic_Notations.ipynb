```python
import random as rnd
import time
import math
import matplotlib.pyplot as plt
from statistics import mean
import numpy
from numpy import arange
from pandas import read_csv
from scipy.optimize import curve_fit
```

```python
inp_sizes = []
for i in range(100):
    inp_sizes.append(10*(i+1))
```

```python
#Bubble sort
def bubble_sort(arr,n):
    i=0

    while(i<n):
        j=0
        while(j<n-1-i):
            
            if(arr[j]>arr[j+1]):
                
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
            j+=1
        i+=1
########
times=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(-10000,10000))
    test_time=[]
    for tests in range(5):
            start=time.time()
            bubble_sort(arr,n)
            test_time.append(time.time()-start)
    times.append(mean(test_time))
########
def quadratic(x, a, b, c):
    return a * x**2 + b * x +  c

def logarithmic(x,a):
    return a*x*numpy.log2(x)

def linear(x,a,b):
    return a*x + b

x,y=inp_sizes,times

########
parameters,_ = curve_fit(quadratic,x,y)
a,b,c = parameters
x1=arange(min(inp_sizes),max(inp_sizes)+10,10)
y1=[]
for j in inp_sizes:
    y1.append(quadratic(j,a,b,c))

########
d,_ = curve_fit(logarithmic,x,y)
x2=x1
y2=[]
for j in inp_sizes:
    y2.append(logarithmic(j,d))

########
parameters,_ = curve_fit(linear,x,y)
e,f=parameters
x3=x1
y3=[]
for j in inp_sizes:
    y3.append(linear(j,e,f))
    
########    
p1=plt.figure(1)   
plt.plot(x1,y1,'--',color = 'red')
plt.title("y=an^2 + bn + c")
plt.scatter(x,y)
########
p2=plt.figure(2)
plt.plot(x2,y2,'--',color='red')
plt.title("y=nlogn")
plt.scatter(x,y)
########
p3=plt.figure(3)
plt.plot(x3,y3,'--',color='red')
plt.title("y=n")
plt.scatter(x,y)

plt.show

```

```python
#Insertion sort
def insertion_sort(arr,n):
    i=0
    while(i<n):
        j=i
        while(j>0):
            if(arr[j]<arr[j-1]):
                temp=arr[j]
                arr[j]=arr[j-1]
                arr[j-1]=temp
            j-=1
        i+=1
########
times=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(-10000,10000))
    test_time=[]
    for tests in range(5):
            start=time.time()
            insertion_sort(arr,n)
            test_time.append(time.time()-start)
    times.append(mean(test_time))
########
def quadratic(x, a, b, c):
    return a * x**2 + b * x +  c

def logarithmic(x,a):
    return a*x*numpy.log2(x)

def linear(x,a,b):
    return a*x + b

x,y=inp_sizes,times

########
parameters,_ = curve_fit(quadratic,x,y)
a,b,c = parameters
x1=arange(min(inp_sizes),max(inp_sizes)+10,10)
y1=[]
for j in inp_sizes:
    y1.append(quadratic(j,a,b,c))

########
d,_ = curve_fit(logarithmic,x,y)
x2=x1
y2=[]
for j in inp_sizes:
    y2.append(logarithmic(j,d))

########
parameters,_ = curve_fit(linear,x,y)
e,f=parameters
x3=x1
y3=[]
for j in inp_sizes:
    y3.append(linear(j,e,f))
    
########    
p1=plt.figure(1)   
plt.plot(x1,y1,'--',color = 'red')
plt.title("y=an^2 + bn + c")
plt.scatter(x,y)
########
p2=plt.figure(2)
plt.plot(x2,y2,'--',color='red')
plt.title("y=nlogn")
plt.scatter(x,y)
########
p3=plt.figure(3)
plt.plot(x3,y3,'--',color='red')
plt.title("y=n")
plt.scatter(x,y)

plt.show

```

```python
#Selection sort
def selection_sort(arr,n):
    for i in range(n):
        min_idx=i
        for j in range(i+1,n):

            if(arr[min_idx]>arr[j]):
                min_idx=j
      
        temp=arr[min_idx]
        arr[min_idx]=arr[i]
        arr[i]=temp
########
times=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(-10000,10000))
    test_time=[]
    for tests in range(5):
            start=time.time()
            selection_sort(arr,n)
            test_time.append(time.time()-start)
    times.append(mean(test_time))
########
def quadratic(x, a, b, c):
    return a * x**2 + b * x +  c

def logarithmic(x,a):
    return a*x*numpy.log2(x)

def linear(x,a,b):
    return a*x + b

x,y=inp_sizes,times

########
parameters,_ = curve_fit(quadratic,x,y)
a,b,c = parameters
x1=arange(min(inp_sizes),max(inp_sizes)+10,10)
y1=[]
for j in inp_sizes:
    y1.append(quadratic(j,a,b,c))

########
d,_ = curve_fit(logarithmic,x,y)
x2=x1
y2=[]
for j in inp_sizes:
    y2.append(logarithmic(j,d))

########
parameters,_ = curve_fit(linear,x,y)
e,f=parameters
x3=x1
y3=[]
for j in inp_sizes:
    y3.append(linear(j,e,f))
    
########    
p1=plt.figure(1)   
plt.plot(x1,y1,'--',color = 'red')
plt.title("y=an^2 + bn + c")
plt.scatter(x,y)
########
p2=plt.figure(2)
plt.plot(x2,y2,'--',color='red')
plt.title("y=nlogn")
plt.scatter(x,y)
########
p3=plt.figure(3)
plt.plot(x3,y3,'--',color='red')
plt.title("y=n")
plt.scatter(x,y)

plt.show
```

```python
#Merge sort
import math
def cmbn(A,B): #Function to combine two sequences(sorted) into a single sorted sequence
    C=[]
    while(len(A)>0 or len(B)>0):
        if(len(A)==0 and len(B)>0):
            C.append(B[0])
            B.pop(0)
        if(len(B)==0 and len(A)>0):
            C.append(A[0])
            A.pop(0)
        if(len(A)>0 and len(B)>0 and A[0]<B[0]): #The smallest between the first elements of A and B is appended to C
            C.append(A[0]) #The element appended is deleted from original array passed to the function
            A.pop(0)
        else:
            if(len(B)>0):
                C.append(B[0])
                B.pop(0)
    return C
def merge_sort(A,l,r):
    if(l<=r):
        if(l==r):
            C=[]
            C.append(A[l])
            return C
        if((l==r-1) and r<len(A)):
            C=[]
            C.append(min(A[l],A[r]))
            C.append(max(A[l],A[r]))
            return C
        m=math.floor(l+(r-l+1)/2)
        #Now the array is divided into two halves, the halves are sorted individually and merged into one single sorted array
        if(m==r):
            B=merge_sort(A,l,m-1) #left half
        else:
            B=merge_sort(A,l,m) 
        C=merge_sort(A,m+1,r)  #right half 
        D=cmbn(B,C) #Calling the combining function
        return D
########
times=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(-10000,10000))
    test_time=[]
    for tests in range(5):
            start=time.time()
            merge_sort(arr,0,n-1) 
            test_time.append(time.time()-start)
    times.append(mean(test_time))
########
def quadratic(x, a):
    return a * x**2 

def logarithmic(x,a):
    return a*x*numpy.log2(x)

def linear(x,a,b):
    return a*x + b

x,y=inp_sizes,times

########
parameters,_ = curve_fit(quadratic,x,y)
a = parameters
x1=arange(min(inp_sizes),max(inp_sizes)+10,10)
y1=[]
for j in inp_sizes:
    y1.append(quadratic(j,a))

########
d,_ = curve_fit(logarithmic,x,y)
x2=x1
y2=[]
for j in inp_sizes:
    y2.append(logarithmic(j,d))

########
parameters,_ = curve_fit(linear,x,y)
e,f=parameters
x3=x1
y3=[]
for j in inp_sizes:
    y3.append(linear(j,e,f))
    
########    
p1=plt.figure(1)   
plt.plot(x1,y1,'--',color = 'red')
plt.title("y=an^2")
plt.scatter(x,y)
########
p2=plt.figure(2)
plt.plot(x2,y2,'--',color='red')
plt.title("y=nlogn")
plt.scatter(x,y)
########
p3=plt.figure(3)
plt.plot(x3,y3,'--',color='red')
plt.title("y=n")
plt.scatter(x,y)

plt.show
```

```python
#Quick sort

def partition(arr,l,r):
    pivot = arr[r] #last element is pivot
    i=l-1
    for j in range(l,r):
        if(arr[j]<pivot): #if element is less than pivot, we increase i and swap arr[i] with arr[j]
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[r]=arr[r],arr[i+1] #finally we place pivot at the right position
    return (i+1)

def quick_sort(arr,l,r):
    if(l<r):
        ind_part = partition(arr,l,r)
        quick_sort(arr,l,ind_part-1)
        quick_sort(arr,ind_part+1,r)
########
times=[]
for n in inp_sizes:
    arr=[]
    for i in range(n):
        arr.append(rnd.randint(-10000,10000))
    test_time=[]
    for tests in range(5):
            start=time.time()
            quick_sort(arr,0,n-1)
            test_time.append(time.time()-start)
    times.append(mean(test_time))
########
def quadratic(x, a):
    return a * x**2 
def logarithmic(x,a):
    return a*x*numpy.log2(x)

def linear(x,a,b):
    return a*x + b

x,y=inp_sizes,times

########
parameters,_ = curve_fit(quadratic,x,y)
a = parameters
x1=arange(min(inp_sizes),max(inp_sizes)+10,10)
y1=[]
for j in inp_sizes:
    y1.append(quadratic(j,a))

########
d,_ = curve_fit(logarithmic,x,y)
x2=x1
y2=[]
for j in inp_sizes:
    y2.append(logarithmic(j,d))

########
parameters,_ = curve_fit(linear,x,y)
e,f=parameters
x3=x1
y3=[]
for j in inp_sizes:
    y3.append(linear(j,e,f))
    
########    
p1=plt.figure(1)   
plt.plot(x1,y1,'--',color = 'red')
plt.title("y=an^2")
plt.scatter(x,y)
########
p2=plt.figure(2)
plt.plot(x2,y2,'--',color='red')
plt.title("y=nlogn")
plt.scatter(x,y)
########
p3=plt.figure(3)
plt.plot(x3,y3,'--',color='red')
plt.title("y=n")
plt.scatter(x,y)

plt.show


```
