```python
import random as rnd
```

```python
def best_dist(n, shortest, blocked, start_x, start_y, end_x, end_y):
    
    if start_x>=n or start_y>=n or blocked[start_x][start_y]: #Checking if the place is out of bounds or is blocked
        return 100000000
    
    if shortest[start_x][start_y] != 100000000: # If the value has already been processed, we return it. Avoids repititions
        return shortest[start_x][start_y]
    
    if start_x == end_x and start_y == end_y: # When we reach the end, returns 0
        return 0
    
    # Considering all 4 possible moves
    s1 = best_dist(n, shortest, blocked, start_x,start_y+1, end_x, end_y) 
    s2 = best_dist(n, shortest, blocked, start_x+1,start_y, end_x, end_y)
    s3 = best_dist(n, shortest, blocked, start_x,start_y-1, end_x, end_y)
    s4 = best_dist(n, shortest, blocked, start_x-1,start_y, end_x, end_y)
    
    min_val = min(s1,s2,s3,s4)
    #if min_val is lesser than previously stored, then we update it
    if(min_val < shortest[start_x][start_y]):
        shortest[start_x][start_y] = min_val + 1 # shortest dist to this point is 1 + shortest dist to previous point
 
    return shortest[start_x][start_y]
    
def path_create(n, shortest, start_x, start_y, end_x, end_y, blocked):
    i = end_x
    j = end_y
    path = [] # path will give the shortest path from start to end
    path.append((end_x,end_y))

    # we iterate i and j until we reach the start, based on best previous place, we update i and j 
    while i!=start_x and j!=start_y and i<n and j<n:
        if(i==n-1):
            s1 = 100000000
        else:
            s1 = shortest[i+1][j]
        if(j==n-1):
            s2 = 100000000
        else:
            s2 = shortest[i][j+1]
        if(i==0):
            s3 = 100000000
        else:
            s3 = shortest[i-1][j]
        if(j==0):
            s4 = 100000000
        else:
            s4 = shortest[i][j-1]

        min_val = min(s1,s2,s3,s4)
        
        if min_val == s1:
            i+=1
            path.append((i,j))
        elif min_val == s2:
            j+=1
            path.append((i,j))
        elif min_val == s3: 
            i-=1
            path.append((i,j))
        elif min_val == s4: 
            j-=1
            path.append((i,j))
            
    path.append((start_x,start_y))
    path.reverse() # Reversing to produce path from start to end
    return path
        
         
    
    
```
